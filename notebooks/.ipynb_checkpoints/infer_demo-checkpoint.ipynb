```python
def load_frames(video_path, start_idx, length=15):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    if start_idx > frame_count:
        print("start_index out of range!!!")
        print('Nothing will be loaded!!!')
        
    if start_idx+length > frame_count:
        print("Warning!!!")
        print('only {} frames has been loaded!!!'.format(frame_count-start_idx))
        
    frames_buffer = []
    
    # set starting point here
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_idx)

    for i in range(length):
        ret, frame = cap.read()

        if not ret:
            break
        # covert into RBG from to correct the image colors
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, [640,360])
        frames_buffer.append(frame)
    return frames_buffer

def make_gif(frames_buffer, fps=10):
    gif_file = BytesIO()
    imageio.mimsave(gif_file, [skimage.img_as_ubyte(frame) for frame in frames_buffer], 'GIF', fps=fps)
    x = Image(data=gif_file.getvalue())
    return x
```

```python
video_fns = [f for f in os.listdir(infer_dir) if not f.startswith('.')]
video_path = os.path.join(infer_dir, video_fns[0])

frames_buffer = load_frames(video_path, 45, length=15)

# these three lines can't write in functions or loops
# and only works in jupyter; or ipython io system
gif = make_gif(frames_buffer, fps=10)
```

```python
plt.figure(figsize=[50,5])
plt.scatter(sf.index.values, sf['seizure_free'], alpha=0.3, color='green')
plt.scatter(t3.index.values, t3['type_3'], alpha=0.3, color='red')
```

```python
def most_frequent(List):
    counter = 0
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num


temp = []
final_labels = []
i = 0

while i < len(G377['label']):
    if len(temp) < 15:
        temp.append(G377['label'].iloc[i])
        i += 1

        
    else:
        final_labels.append(most_frequent(temp))
        temp = []
        i -= 14
                
```

```python
smoothed = G377[:len(final_labels)].copy()
```

```python
smoothed['final'] = final_labels
```

```python
t3 = smoothed[smoothed['final'] == 'type_3']
sf = smoothed[smoothed['final'] == 'seizure_free']

plt.figure(figsize=[50,5])
plt.scatter(sf.index.values, sf['seizure_free'], alpha=0.3, color='green')
plt.scatter(t3.index.values, t3['type_3'], alpha=0.3, color='red')
```
