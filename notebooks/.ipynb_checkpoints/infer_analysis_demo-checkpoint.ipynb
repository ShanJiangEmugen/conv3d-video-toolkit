```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def most_frequent(List):
    counter = 0
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num
```

```python
G377 = pd.read_csv('predictions/G377_prediction.csv', index_col=0)
G377['label'] = G377.idxmax(axis=1)
_, counts = np.unique(G377['label'], return_counts=True)
```

```python
t3 = G377[G377['label'] == 'type_3']
sf = G377[G377['label'] == 'seizure_free']
```

```python
plt.figure(figsize=[100,5])
plt.scatter(sf.index.values, sf['seizure_free'], alpha=0.6, color='green', marker='|')
plt.scatter(t3.index.values, t3['type_3'], alpha=0.6, color='red', marker='|')
```

```python
temp = []
final_labels = []
i = 0

while i < len(G377['label']):
    if len(temp) < 15:
        temp.append(G377['label'].iloc[i])
        i += 1

        
    else:
        final_labels.append(most_frequent(temp))
        temp = []
        i -= 14
                
smoothed = G377[:len(final_labels)].copy()
smoothed['final'] = final_labels
```

```python
t3 = smoothed[smoothed['final'] == 'type_3']
sf = smoothed[smoothed['final'] == 'seizure_free']

plt.figure(figsize=[100,5])
plt.scatter(sf.index.values, sf['seizure_free'], alpha=0.6, color='green', marker='|')
plt.scatter(t3.index.values, t3['type_3'], alpha=0.6, color='red', marker='|')
```
